```
''' 2020 Nov 3rd Leetcode challenge  Consecutive Characters '''
# my solution linear search, faster than 99% 
class Solution:
    def maxPower(self, s: str) -> int:
        max_len = 1
        prev_c = s[0]
        tmp_len = 1
        for c in s[1:]:
            if c == prev_c:
                tmp_len += 1
                if tmp_len > max_len:
                    max_len = tmp_len
            else:
                tmp_len = 1
                
            prev_c = c       
        return max_len


# fastest solution 
class Solution:
    def maxPower(self, s):
        s = chain(s, '\n')
        res = cur = 1
        prev = next(s)
        for x in s:
            if x != prev:
                if cur > res:
                    res = cur
                cur = 1
            else:
                cur += 1
            prev = x
        return res
```
