```python
# 運動習慣
# Programming challenge description:
# 太郎くんは曜日ごとに決まった時間だけ運動をすることにしています。

# 太郎くんの曜日ごとの運動時間の長さが与えられるので、連続する 2 日の運動時間の長さの差の絶対値が最大となるときのその値を求めてください。

# Input:
# 標準入力に以下の形式で 7 つの整数が空白区切りで与えられます。 入力データの行末に改行があることに注意してください。

# T_mon T_tue T_wed T_thu T_fri T_sat T_sun
# T_mon ... 月曜日の運動時間(分)
# T_tue ... 火曜日の運動時間(分)
# T_wed ... 水曜日の運動時間(分)
# T_thu ... 木曜日の運動時間(分)
# T_fri ... 金曜日の運動時間(分)
# T_sat ... 土曜日の運動時間(分)
# T_sun ... 日曜日の運動時間(分)
# 制約
# 与えられる値は全て 0 以上 100 以下の整数
# Output:
# 連続する 2 日の運動時間の長さの差の最大値を標準出力に出力してください。 出力は概ね 10 秒以内に得られるようにしてください。処理時間が長すぎる場合は得点が得られません。

# 入出力のサンプル
# 例1
# 入力

# 20 30 20 40 30 10 20
# 出力

# 20
# 月曜日と火曜日の差の絶対値は 10 分
# 火曜日と水曜日の差の絶対値は 10 分
# 水曜日と木曜日の差の絶対値は 20 分
# 木曜日と金曜日の差の絶対値は 10 分
# 金曜日と土曜日の差の絶対値は 20 分
# 土曜日と日曜日の差の絶対値は 10 分
# 日曜日と月曜日の差の絶対値は 0 分
# この中から、最も大きい 20 (分)を出力してください。

# 例2
# 入力

# 15 20 25 30 35 40 45
# 出力

# 30
# 日曜日と月曜日の差の絶対値が 30 分で最大です。

# Test 1
# Test Input
# Download Test 1 Input
# 100 23 89 1 87 54 0
# Expected Output
# Download Test 1 Input
# 100
# Test 2
# Test Input
# Download Test 2 Input
# 15 20 25 30 35 40 45
# Expected Output
# Download Test 2 Input
# 30
# Test 3
# Test Input
# Download Test 3 Input
# 20 30 20 40 30 10 20
# Expected Output
# Download Test 3 Input
# 20
# Test 4
# Test Input
# Download Test 4 Input
# 32 32 54 12 52 56 8
# Expected Output
# Download Test 4 Input
# 48
# question 1
import sys
for line in sys.stdin:
    values = [int(x) for x in line.split()]
    max_value = abs(values[0] - values[len(values) - 1])
    for i in range(1, len(values)):
        max_value = max(abs(values[i] - values[i-1]), max_value)
    print(max_value,end="")
```

```python
# 読書習慣
# Programming challenge description:
# 太郎くんは読書を習慣づけることにしました。 同じ本を読んでばかりいては飽きてしまうので、たまに新しい本を買っています。 太郎くんは本の面白さを数値化していて、持っている本のうちできるだけ面白い（面白さの数値が高い）本を読んでいます。

# 太郎くんの N 日間の行動記録が与えられます。太郎くんの 1 日にとる行動は以下のいずれかです。

# buy x : 面白さが x の本を 1 冊買い、買った本を読みます。
# read : 持っている本のうち、直近 K 日に読んだことのあるものを除いた中からもっとも面白い本を読みます。該当する本が複数ある場合は、どれを読んでもよいです。直近 K 日とは、K 日前から前日にかけての K 日間のことを指します。
# 1 日目の開始時点で、持っている本は 0 冊です。 それぞれの日の太郎くんが読んだ本の面白さを求めてください。

# Input:
# 標準入力に以下の形式で与えられます。 入力データの行末に改行があることと、複数行で入力が与えられることに注意してください。

# N K
# q_1
# q_2
# ...
# q_N
# N ... 行動記録の日数
# K ... 太郎くんが同じ本を読むことができない日数
# q_i ... i 日目の行動記録
# 制約
# 入力で与えられる数は全て整数
# 4 <= N <= 200
# 1 <= K <= N
# q_i は「 buy x, read 」のいずれか（x は整数）
# 1 <= x <= 100
# read が記録されている日において、読むことのできる本が必ず存在することが保証されます。
# Output:
# N 日間に読んだ本の面白さを 1 日目から順に 1 行ずつ出力してください。

# 出力は概ね 10 秒以内に得られるようにしてください。処理時間が長すぎる場合は得点が得られません。

# 入出力のサンプル
# 例 1
# 入力

# 5 1
# buy 2
# buy 3
# buy 1
# read
# read
# 出力

# 2
# 3
# 1
# 3
# 2
# 1 日目 (buy 2): 面白さ 2 の本を買い、買った本を読みます。
# 2 日目 (buy 3): 面白さ 3 の本を買い、買った本を読みます。
# 3 日目 (buy 1): 面白さ 1 の本を買い、買った本を読みます。
# 4 日目 (read): 面白さ 3 の本を読みます。
# 5 日目 (read): 面白さ 2 の本を読みます。
# 5 日目において、「2, 3, 1」の本を太郎くんは持っていますが、4 日目に面白さ 3 の本を読んでいるため面白さ 3 の本は選べないことに注意してください。 以上より、各日に読んだ本の面白さである「2, 3, 1, 3, 2」 を 1 行ずつ出力してください。

# 例 2
# 入力

# 5 2
# buy 2
# buy 3
# buy 1
# read
# read
# 出力

# 2
# 3
# 1
# 2
# 3
# 1 日目 (buy 2): 面白さ 2 の本を買い、買った本を読みます。
# 2 日目 (buy 3): 面白さ 3 の本を買い、買った本を読みます。
# 3 日目 (buy 1): 面白さ 1 の本を買い、買った本を読みます。
# 4 日目 (read): 面白さ 2 の本を読みます。
# 5 日目 (read): 面白さ 3 の本を読みます。
# 4 日目において、面白さ 1 の本と面白さ 3 の本が選べないことに注意してください。

# 例 3
# 入力

# 8 2
# buy 2
# buy 2
# buy 1
# read
# buy 3
# read
# read
# read
# 出力

# 2
# 2
# 1
# 2
# 3
# 2
# 2
# 3
# 6 日目には 2 日目に買った面白さ 2 の本を、7 日目には 1 日目に買った面白さ 2 の本を読んでいることに注意してください。

# Test 1
# Test Input
# Download Test 1 Input
# 5 1
# buy 2
# buy 3
# buy 1
# read
# read
# Expected Output
# Download Test 1 Input
# 2
# 3
# 1
# 3
# 2
# Test 2
# Test Input
# Download Test 2 Input
# 5 2
# buy 2
# buy 3
# buy 1
# read
# read
# Expected Output
# Download Test 2 Input
# 2
# 3
# 1
# 2
# 3
# Test 3
# Test Input
# Download Test 3 Input
# 8 2
# buy 2
# buy 2
# buy 1
# read
# buy 3
# read
# read
# read
# Expected Output
# Download Test 3 Input
# 2
# 2
# 1
# 2
# 3
# 2
# 2
# 3
# Test 4
# Test Input
# Download Test 4 Input
# 12 3
# buy 100
# buy 100
# buy 100
# buy 100
# buy 99
# read
# read
# read
# buy 99
# buy 98
# read
# read
# Expected Output
# Download Test 4 Input
# 100
# 100
# 100
# 100
# 99
# 100
# 100
# 100
# 99
# 98
# 100
# 100
# question 2
import sys
line_count = 0
buy_arr = []
read_arr = []
for line in sys.stdin:
    if line_count == 0:
        n, k = (int(x) for x in line.split())
    else:
        if 'buy' in line:
            value = int(line[4:])
            buy_arr.append(value)
            read_arr.append(value)
        elif 'read' in line:
            acceptable = read_arr[: -k]
            m = max(acceptable)
            buy_arr.append(m)
            read_arr.pop(read_arr.index(m))
            read_arr.append(m)
            print(acceptable)
    line_count += 1

for i in buy_arr:
    print(i)
```

```python
# カード
# Programming challenge description:
# 片面が赤、もう一方の面が青に塗られているカードが N 枚あり、テーブルの上に一列に並べられています。

# 現在、カードの表面の色が左から順番に A_1...A_N となるように並んでいます。

# あなたはこのカードの列に操作を行うことで、カードの表面の色が左から順番に B_1...B_N となるようにしたいです。

# あなたができる操作は以下の 2 種類で、どちらの操作も何回でも好きな回数だけ行うことができますが、操作 1 回ごとにコストがかかります。

# 隣り合う 2 枚のカードの位置を入れ替える: コスト X 円
# カードを 1 枚選び、裏返す: コスト Y 円
# 目標を達成するために必要なコストの合計値の最小値は何円でしょうか？

# Input:
# 標準入力に以下の形式で与えられます。 入力データの行末に改行があることに注意してください。

# N X Y
# A_1 A_2 ... A_N
# B_1 B_2 ... B_N
# N ... カードの枚数
# X ... 隣り合う 2 枚のカードを入れ替えるコスト(円)
# Y ... カードを 1 枚選び裏返すコスト(円)
# A_i ... 初期状態の左から i 番目のカードの表面の色。この値が R の場合は赤い面が表になっており、B の場合は青い面が表になっていることを表す。
# B_i ... 目標状態の左から i 番目のカードの表面の色。この値が R の場合は赤い面が表になっており、B の場合は青い面が表になっていることを表す。
# 制約
# 入力で与えられる数は全て整数
# 2 <= N <= 16
# 0 <= X <= 100
# 0 <= Y <= 100
# A_i および B_i は R か B のいずれか
# Output:
# N 枚のカードの表面の色が指定された順番に並ぶようにするためにかかるコストの最小値(円)を標準出力に出力してください。 出力は概ね 10 秒以内に得られるようにしてください。処理時間が長すぎる場合は得点が得られません。

# 入出力のサンプル
# 例 1
# 入力

# 3 3 5
# R R B
# B B R
# 出力

# 8
# 説明

# 以下の手順がコストの合計値が最小となる操作の一例です。

# 2 枚目と 3 枚目を入れ替える (3 円)
# 1 枚目を裏返す (5 円)
# このとき 8 円で目標の色の並びにすることができます。また、7 円以下で達成することはできないので、出力するべき答は 8 円になります。

# なお、3 枚全てを裏返すことでも目標の色の並びを達成できますが、この場合は 9 円かかるため最適ではありません。

# 例 2
# 入力

# 5 13 7
# R B R R B
# B R B R R
# 出力

# 27
# 説明

# 以下の手順がコストの合計値が最小となる操作の一例です。

# 1 枚目を裏返す (7 円)
# 2 枚目と 3 枚目を入れ替える (13 円)
# 5 枚目を裏返す (7 円)
# このとき合計 27 円で目標の色の並びにすることができます。

# 例 3
# 入力

# 16 73 59
# B B R R R B R R B R B R B B R B
# R R B R R B B B R B B B R R R R
# 出力

# 514
# Test 1
# Test Input
# Download Test 1 Input
# 3 3 5
# R R B
# B B R
# Expected Output
# Download Test 1 Input
# 8
# Test 2
# Test Input
# Download Test 2 Input
# 5 13 7
# R B R R B
# B R B R R
# Expected Output
# Download Test 2 Input
# 27
# Test 3
# Test Input
# Download Test 3 Input
# 16 73 59
# B B R R R B R R B R B R B B R B
# R R B R R B B B R B B B R R R R
# Expected Output
# Download Test 3 Input
# 514
# Test 4
# Test Input
# Download Test 4 Input
# 16 100 100
# B B B B B B B B B B B B B B B B
# R R R R R R R R R R R R R R R R
# Expected Output
# Download Test 4 Input
# 1600
# question 3
import sys
lines = sys.stdin.readlines()

n, x, y = (int(x) for x in lines[0].split())
target = lines[1].split()
current = lines[2].split()


sum_count = 0
for i in range(len(current) - 1):
    # print("--------------")
    # print(target)
    # print(current)
    if current[i] != target[i]:
        if x <= 2 * y and current[i + 1] == target[i] and current[i] == target[i + 1]:
            # Choose swap
            sum_count += x
            # print('swap', x)
            current[i], current[i + 1] = current[i + 1], current[i]
        else:
            # Choose turn
            sum_count += y
            # print('turn', y)
            current[i] = target[i]

# print("--------------")
# print(target)
# print(current)
if current[len(current) - 1] != target[len(current) - 1]:
    # print('turn', y)
    sum_count += y

print(sum_count)
```

```python

def validateAddresses(addresses):
    # Write your code here
    ret = []
    for address in addresses:
        if address.count('.') != 3:
            ret.append('NO')
            continue
        
        nums = address.split('.')
        is_valid = True
    
        for num in nums:
            try:  
                if num[0] == '0':
                    num = int(num, 8)
                else:
                    num = int(num)
            
                if num < 0 or num > 255:
                    is_valid = False
            except Exception:
                is_valid = False
            
        if is_valid == False:
            ret.append('NO')
        else:
            ret.append('IPv4')
            
    return ret    
        
        
```
