```
''' Leetcode Max Number of K-Sum Pairs
You are given an integer array nums and an integer k.
In one operation, you can pick two numbers from the array whose sum equals k and remove them from the array.
Return the maximum number of operations you can perform on the array.

Example 1:

Input: nums = [1,2,3,4], k = 5
Output: 2
Explanation: Starting with nums = [1,2,3,4]:
- Remove numbers 1 and 4, then nums = [2,3]
- Remove numbers 2 and 3, then nums = []
There are no more pairs that sum up to 5, hence a total of 2 operations

'''
# my submission 40%

class Solution:
    def maxOperations(self, nums: List[int], k: int) -> int:
        ret = 0
        my_dic = Counter(nums)
        
        for x in nums:
            # frequency of x > 0
            if x in my_dic and my_dic[x] > 0:
                y = k - x
                if y == x:
                    ret += my_dic[x]//2
                    my_dic[x] = my_dic[x] %2
                else:
                    # y != x and frequency of y > 0
                    if y in my_dic and my_dic[y] > 0:
                        # found match x,y
                        tmp = min(my_dic[x], my_dic[y])
                        my_dic[x] -= tmp
                        my_dic[y] -= tmp
                        # remove pair x,y
                        ret += tmp
        
        return ret


# fastest solution 

class Solution:
    def maxOperations(self, nums: List[int], k: int) -> int:
        counts = collections.Counter(nums)
        res = 0
        for i in counts:
            if i * 2 == k:
                res += math.floor(counts[i]/2)
            elif i <= k/2:
                if k - i in counts:
                    res += min(counts[i],counts[k-i])
        return res
```

```
# my solution 
# using extra helper
class Solution:
    def longestPalindrome(self, s: str) -> str:
        max_pal = ""
        def helper(s, l, r):
            while l>=0 and r<len(s) and (s[l] == s[r]):
                l, r = l-1, r+1
            return s[l+1:r]
        
        for i in range(len(s)):
            max_pal = max([helper(s,i,i), helper(s,i,i+1), max_pal], key=len)
            
        return max_pal


# fastest solution using loop
class Solution:
    
    def longestPalindrome(self, string: str) -> str:
        if len(string) <= 1 or string == string[::-1]:
            return string

        start = 0
        length = 1

        for i in range(len(string)):
            odd = string[i - length - 1 : i + 1]
            even = string[i - length : i + 1]
            
            if 0 <= i - length - 1 and odd == odd[::-1]:
                start = i - length - 1
                length += 2
            elif 0 <= i - length and even == even[::-1]:
                start = i - length
                length += 1

        return string[start : start + length]
```

```
""" Leetcode Jan 23 2021 """"
# my solution beats 89% 
class Solution:
    def diagonalSort(self, mat: List[List[int]]) -> List[List[int]]:
        rows, cols = len(mat), len(mat[0])
        # 1st iteration (the first half)
        for i in range(rows):
            j, temp_diag, pos = 0, [], []
            while i < rows and j < cols:
                temp_diag.append(mat[i][j])
                pos.append([i,j])
                i, j = i+1, j+1
            temp_diag.sort() 
            for index, p in enumerate(pos):
                mat[p[0]][p[1]] = temp_diag[index]
                
        # 2nd iteration (the second half)
        for j in range(1,cols):
            i, temp_diag, pos = 0, [], []
            while i < rows and j < cols:
                temp_diag.append(mat[i][j])
                pos.append([i,j])
                i, j = i+1, j+1
            temp_diag.sort() 
            for index, p in enumerate(pos):
                mat[p[0]][p[1]] = temp_diag[index]
                
        return mat 

# fastest submission 
"""
the dict before sorting looks like below
defaultdict(<class 'list'>, 
{0: [3, 2, 1], -1: [3, 1, 2], -2: [1, 2], -3: [1], 1: [2, 1], 2: [1]})

sort the dict values, then put them back
"""
class Solution:
    def diagonalSort(self, A: List[List[int]]) -> List[List[int]]:
        n, m = len(A), len(A[0])
        d = collections.defaultdict(list)
        # save the diagonal list to dict
        for i in range(n):
            for j in range(m):
                d[i - j].append(A[i][j])
        
        # sort all dict values
        for k in d:
            d[k].sort(reverse=1)
        
        # put them back to the origional matrix
        for i in range(n):
            for j in range(m):
                A[i][j] = d[i - j].pop()
        return A

```

```
# LEETCODE solution 1:
class Solution:
    def diagonalSort(self, mat: List[List[int]]) -> List[List[int]]:
        # hashmap to keep the diagonals
        h = defaultdict(list)
        
        # fill the hashmap
        n, m = len(mat), len(mat[0])
        for i in range(n):
            for j in range(m):
                heappush(h[i - j], mat[i][j])

        # build output
        for i in range(n):
            for j in range(m):
                mat[i][j] = heappop(h[i - j])
        
        return mat
```

```
''' Merge k Sorted Lists '''
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeKLists(self, lists: List[ListNode]) -> ListNode:
        import heapq
        cur = head = ListNode(0)
        queue = []
        count = 0
        for l in lists:
            if l is not None:
                count += 1
                heapq.heappush(queue, (l.val, count, l))
        while len(queue) > 0:
            _, _, cur.next = heapq.heappop(queue)
            cur = cur.next
            if cur.next is not None:
                count += 1
                heapq.heappush(queue, (cur.next.val, count, cur.next))
        return head.next
```

```
'''Check If All 1's Are at Least Length K Places Away
Given an array nums of 0s and 1s and an integer k, return True if all 1's are at least k places away from each other, otherwise return False.

Example 1:
Input: nums = [1,0,0,0,1,0,0,1], k = 2
Output: true
Explanation: Each of the 1s are at least 2 places away from each other.
'''
# my solution 1: using stack
class Solution:
    def kLengthApart(self, nums: List[int], k: int) -> bool:
        stk = [ArithmeticError]
        for i in range(len(nums)):
            if nums[i] == 1:
                if stk and i - stk[-1] - 1 < k:
                    return False
                stk.append(i)
        return True 


# my solution 2: using memo var
class Solution:
    def kLengthApart(self, nums: List[int], k: int) -> bool:
        memo = -1
        for i in range(len(nums)):
            if nums[i] == 1:
                if memo > -1 and i - memo - 1 < k:
                    return False
                memo = i
        return True 
```
