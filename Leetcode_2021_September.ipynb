```python
# my solution 
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        memo = []
        while head:
            memo.append(head.val)
            head = head.next
        
        cur = dummy = ListNode(0)
        while memo:
            cur.next = ListNode(memo.pop())
            cur = cur.next
        
        return dummy.next


# fastest submission
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if head == None:
            return None
        
        previousNode = None
        currentNode = head
        
        while True:
            # Swap the direction of the next pointer to point to the previous
            nextNode = currentNode.next
            currentNode.next = previousNode
            
            # Move onto the next
            if nextNode is None:
                break
            
            previousNode = currentNode
            currentNode = nextNode
            
        return currentNode     
```

```python
### my solution 32 ms

class Solution:

    from itertools import combinations
    def collinear(self, points):
        x1, y1, x2, y2, x3, y3 = points[0][0], points[0][1], points[1][0], points[1][1],points[2][0], points[2][1] 
        if ((y3 - y2)*(x2 - x1) == (y2 - y1)*(x3 - x2)):
            return True
        else:
            return False

    def tictactoe(self, moves: List[List[int]]) -> str:
        l, A_lst, B_lst = len(moves), moves[0::2], moves[1::2]
            
        for i in combinations(A_lst,3):
            if self.collinear(i):
                    return "A"
        for i in combinations(B_lst,3):
            if self.collinear(i):
                    return "B"
        
        return "Draw" if l == 9 else "Pending"

# 20 ms, fastest solution 

def tictactoe(self, moves: List[List[int]]) -> str:
        A=[0]*8
        B=[0]*8
        for i in range(len(moves)):
            r,c=moves[i]
            player = A if i%2==0 else B
            player[r] += 1
            player[c+3] += 1
            if r==c:
                player[6] += 1
            if r==2-c:
                player[7] += 1
        for i in range(8):
            if A[i]==3:
                return "A"
            if B[i]==3:
                return "B"
        
        return "Draw" if len(moves) == 9 else "Pending"
```
