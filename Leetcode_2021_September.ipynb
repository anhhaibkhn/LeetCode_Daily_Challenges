```python
# my solution 
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        memo = []
        while head:
            memo.append(head.val)
            head = head.next
        
        cur = dummy = ListNode(0)
        while memo:
            cur.next = ListNode(memo.pop())
            cur = cur.next
        
        return dummy.next


# fastest submission
class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if head == None:
            return None
        
        previousNode = None
        currentNode = head
        
        while True:
            # Swap the direction of the next pointer to point to the previous
            nextNode = currentNode.next
            currentNode.next = previousNode
            
            # Move onto the next
            if nextNode is None:
                break
            
            previousNode = currentNode
            currentNode = nextNode
            
        return currentNode     
```
